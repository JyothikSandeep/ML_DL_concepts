Linear Regression is not suitable for classification because it predicts continuous values, while classification requires discrete classes (e.g., 0 or 1).

Example

Suppose we want to classify whether a student passes:

Hours Studied	Pass
1	0
2	0
3	0
4	1
5	1


A linear regression model might learn:

y=0.3x−0.8

Now for:

x = 2 → y = -0.2
x = 4 → y = 0.4
x = 7 → y = 1.3

Problems
1. Predictions can be less than 0 or greater than 1

For classification, probabilities should be between 0 and 1.

Linear regression can predict:

-0.5
1.7
2.3

These are not valid probabilities.

2. Sensitive to outliers

A few extreme points can significantly change the regression line and affect classification decisions.

3. Decision boundary is not probabilistic

Classification often needs:

"What is the probability that this belongs to class 1?"

Linear regression doesn't naturally provide valid probabilities.

Why Logistic Regression Works

Instead of a straight line, logistic regression applies a sigmoid function:

p=
1+e
−z
1
	​


This converts any number into a value between 0 and 1.

Example:

z = -10 → p ≈ 0
z = 0 → p = 0.5
z = 10 → p ≈ 1

Then:

p ≥ 0.5 → Class 1
p < 0.5 → Class 0
Visual Intuition

Linear Regression

Class 0    Class 1

0----0----1----1

---------/---------
        line

Predictions can go below 0 or above 1.

Logistic Regression

Probability

1.0 |        ______
    |      /
0.5 |----/
    |  /
0.0 |/
    +----------------

The S-shaped curve keeps probabilities between 0 and 1.

Interview Answer

Linear Regression is not used for classification because it predicts continuous values and can produce outputs outside the valid probability range [0,1]. Logistic Regression uses a sigmoid function to convert predictions into probabilities between 0 and 1, making it suitable for classification tasks.

Performance Metrics:

1. confusion matrix
2. Accuracy
3. Precision
4. Recall
5. F_Beta score

A Confusion Matrix is used to evaluate how well a Logistic Regression classifier performs.

For a binary classification problem, it looks like this:

| Actual \ Predicted | Positive (1)        | Negative (0)        |
| ------------------ | ------------------- | ------------------- |
| Positive (1)       | True Positive (TP)  | False Negative (FN) |
| Negative (0)       | False Positive (FP) | True Negative (TN)  |


Example

Suppose we have the following confusion matrix:

| Actual \ Predicted | 1  | 0  |
| ------------------ | -- | -- |
| 1                  | 50 | 10 |
| 0                  | 5  | 35 |
Metrics from Confusion Matrix
Accuracy
Accuracy=TP+TN+FP+FN//TP+TN
	​

Accuracy=
50+35+5+10//50+35=100//85=85%
Precision
Precision=
TP+FP//TP
	​

Precision=
50+5//50=0.909

Recall (Sensitivity)
Recall=
TP+FN//TP
	​

Recall=
50+10//50=0.833

F1 Score
F1=Precision+Recall//2×Precision×Recall
	​

F1=0.87

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification

```

```python
# Creating dataset

X,y=make_classification( n_samples=1000,n_features=10,n_classes=2,random_state=42)

X.shape,y.shape
```

```python
X=pd.DataFrame(X)
y=pd.DataFrame(y)

```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)

y_pred
```

```python
# Performance metircs

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

```

```python
score=accuracy_score(y_test,y_pred)

score
```

```python
cm=confusion_matrix(y_test,y_pred)

cm
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

params = {
    'penalty': ['l1', 'l2','elasticnet'],
    'C': [100, 10, 1, 0.1, 0.01],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}

cv = StratifiedKFold()

logistic = LogisticRegression()

model = GridSearchCV(
    estimator=logistic,
    param_grid=params,
    scoring='accuracy',
    cv=cv,
    n_jobs=-1
)

model.fit(X_train, y_train)

print("Best Parameters:", model.best_params_)
print("Best Accuracy:", model.best_score_)
```

```python
model
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report


y_pred=model.predict(X_test)

accuracy=accuracy_score(y_test,y_pred)

cm=confusion_matrix(y_test,y_pred)

report=classification_report(y_test,y_pred)

print(accuracy)
print(cm)
print(report)
```
