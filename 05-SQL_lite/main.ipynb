```python
import sqlite3
```

```python
connection=sqlite3.connect('example.db')
```

```python
cursor=connection.cursor()

cursor.execute('''
 create table if not exists employees(
 id integer primary key,
 name text not null,
 age integer not null,
 profession text not null
 ) 
''')
```

```python
cursor.execute('''
select * from employees
''')
```

```python
cursor.execute('''
insert into employees(id,name,age,profession) values(1,"sandy",12,"engineer")
''')

connection.commit()
```

```python
cursor.execute('''

select * from employees

''')

rows=cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
    update employees set profession="engine" where id=1 
    ''')

connection.commit()
```

```python
cursor.execute('''

select * from employees

''')

rows=cursor.fetchall()

for row in rows:
    print(row)
```
