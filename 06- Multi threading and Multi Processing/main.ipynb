```python
import time
import threading
```

```python
def print_numbers():

    for i in range(5):
        print(i)

def print_alphabets():
    for i in "aeiou":
        print(i)

t=time.time()

print_numbers()
print_alphabets()

finished_time=time.time()-t

print(finished_time)

# 
```

```python
# Addding sleep time

def print_numbers():

    for i in range(5):
        time.sleep(2)
        print(i)

def print_alphabets():

    for i in "aeiou":
        time.sleep(2)
        print(i)

t=time.time()

print_numbers()
print_alphabets()

finished_time=time.time()-t

print(finished_time)

```

```python
# Using multi threading concept
def print_numbers():

    for i in range(5):
        time.sleep(2)
        print(i)

def print_alphabets():

    for i in "aeiou":
        time.sleep(2)
        print(i)

t=time.time()

t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_alphabets)
# Starting threads
t1.start()
t2.start()

finished_time=time.time()-t


print(finished_time)


```

```python
# Multi processing-> We can assign our task like highly computational ones to different process like different cores


import multiprocessing

```

```python
# Using multi threading concept
def print_numbers():

    for i in range(5):
        time.sleep(1)
        print(i)

def print_alphabets():

    for i in "aeiou":
        time.sleep(1.5)
        print(i)


p1=multiprocessing.Process(target=print_numbers())
p2=multiprocessing.Process(target=print_alphabets())
# Starting threads
t=time.time()
p1.start()
p2.start()

finished_time=time.time()-t


print(finished_time)


```
