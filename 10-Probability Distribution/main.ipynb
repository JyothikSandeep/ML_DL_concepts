✅ 2. PDF (Continuous Example)
🌡 Example: Temperature
Let:
X = temperature between 20°C and 30°C

We cannot say:
P(X = 25) → 0 (always zero!)


Instead:
We calculate probability over a range:
P(22<X<26)P(22 < X < 26)P(22<X<26)

✅ Uniform Distribution Example
If all values are equally likely:
f(x)=110,(20<x<30)f(x) = \frac{1}{10}, \quad (20 < x < 30)f(x)=101​,(20<x<30)

📌 Find Probability
P(22<X<26)P(22 < X < 26)P(22<X<26)
=(26−22)×110=410=0.4= (26 - 22) \times \frac{1}{10} = \frac{4}{10} = 0.4=(26−22)×101​=104​=0.4

👉 This is PDF because we use area (integration concept).

✅ 3. CDF (Cumulative Example)

✅ CDF:
F(x)=P(X≤x)F(x) = P(X \le x)F(x)=P(X≤x)

✅ 🔥 Super Simple Intuition
🍫 Chocolate Box Example

```python
# Probability mass function

import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5,6])

pmf=np.ones(6)/6

print(pmf)

plt.bar(x,pmf)
plt.xlabel("X Dice values")
plt.ylabel("PMF")

plt.title("PMF - Dice")

plt.show()
```

```python
# Probability density function

x=np.linspace(20,30,100)

y = np.ones_like(x) * (1 / (30 - 20))


plt.plot(x, y)
plt.xlabel("X (Temperature)")
plt.ylabel("f(x)")
plt.title("PDF - Uniform Distribution")
plt.show()


y
```

```python
#Continous discreate function
x = np.array([1, 2, 3, 4, 5, 6])
pmf = np.ones(6) / 6

cdf=np.cumsum(pmf)

plt.plot(x,cdf)

plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(20, 30, 100)

# CDF formula
cdf = (x - 20) / 10

plt.plot(x, cdf)
plt.xlabel("X")
plt.ylabel("F(x)")
plt.title("CDF - Continuous Uniform Distribution")
plt.show()
```
