```python
# Probability mass function

import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5,6])

pmf=np.ones(6)/6

print(pmf)

plt.bar(x,pmf)
plt.xlabel("X Dice values")
plt.ylabel("PMF")

plt.title("PMF - Dice")

plt.show()
```

```python
# Probability density function

x=np.linspace(20,30,100)

y = np.ones_like(x) * (1 / (30 - 20))


plt.plot(x, y)
plt.xlabel("X (Temperature)")
plt.ylabel("f(x)")
plt.title("PDF - Uniform Distribution")
plt.show()


y
```

```python
#Continous discreate function
x = np.array([1, 2, 3, 4, 5, 6])
pmf = np.ones(6) / 6

cdf=np.cumsum(pmf)

plt.plot(x,cdf)

plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(20, 30, 100)

# CDF formula
cdf = (x - 20) / 10

plt.plot(x, cdf)
plt.xlabel("X")
plt.ylabel("F(x)")
plt.title("CDF - Continuous Uniform Distribution")
plt.show()
```

## Binomial distribution

1. Core Idea

Bernoulli:

one trial

Binomial:

multiple Bernoulli trials
Example

Suppose:

Toss coin 10 times
Count number of heads

Possible outputs:

0,1,2,3,...,10

This is Binomial Distribution.

2. Conditions for Binomial Distribution

A random experiment is Binomial if:

Condition 1 — Fixed Number of Trials

Example:

exactly 10 tosses
Condition 2 — Only Two Outcomes

Each trial:

success/failure
yes/no
0/1
Condition 3 — Same Probability

Probability remains constant.

Example:

coin probability always 0.5
Condition 4 — Independent Trials

One trial does not affect another.

3. Parameters

Binomial has two parameters:

| Parameter | Meaning                |
| --------- | ---------------------- |
| (n)       | number of trials       |
| (p)       | probability of success |


Notation:

X∼Bin(n,p)
Example
X∼Bin(10,0.5)

means:

10 trials
probability of success = 0.5
