```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s = Student("John", 22)

print(s.name)
print(s.age)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return self.name
s=Student("sandy",10)
print(s)
```

```python
class Student:
    def __repr__(self):
        return "Student Object"

s = Student()

print(repr(s))
```

```python
class Student:
    def __str__(self):
        return "Readable"

    def __repr__(self):
        return "Developer"

s = Student()

print(str(s))
print(repr(s))
```

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)

t = Team(["A", "B", "C"])

print(len(t))
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

a = Number(10)
b = Number(20)

print(a + b)
```

```python
class Person:
    def __init__(self, age):
        self.age = age

    def __eq__(self, other):
        return self.age == other.age

p1 = Person(25)
p2 = Person(25)

print(p1 == p2)
```

```python
class Numbers:
    def __init__(self):
        self.data = [10, 20, 30]

    def __getitem__(self, index):
        return self.data[index]

n = Numbers()

print(n[1])
```
