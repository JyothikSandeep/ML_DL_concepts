Statistics is a branch of mathematics that deals with collecting, organizing, analyzing, and interpreting data.

✅ Simple Definition
👉 Statistics = Study of data
It helps us understand patterns, trends, and make decisions based on data.

Suppose you ask:

“What is the average salary in my company?”
“How many people passed an exam?”
“What is the most popular product?”

ypes of Statistics
1. Descriptive Statistics

Summarizes data
Example:

Mean (average)
Median
Mode
Charts, graphs



👉 Example: Average marks of a class

2. Inferential Statistics

Makes predictions or conclusions based on data
Example:

Forecasting
Probability
Hypothesis testing



👉 Example: Predicting election results using sample data

✅ Why Statistics is Important

📊 Helps in decision making
📈 Used in business, IT, data science
🧪 Used in research and experiments
📉 Helps find trends and pattern

measure of central tendency:
1. Mean
2. Median
3. Mode

Measure of dispersion:
1. Variance
2. Standard deviation



```python
import numpy as np
l1=[1,3,5,6]
print(np.mean(l1))
# print(np.mode(l1))
print(np.median(l1))

np.var(l1)
```
